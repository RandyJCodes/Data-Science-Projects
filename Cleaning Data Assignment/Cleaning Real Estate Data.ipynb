```python
houseAge = data.houseAge
```

```python
norm = (houseAge - houseAge.mean()) / houseAge.std() #Standardized equation
norm
```

```python
minmax = (houseAge - houseAge.min()) / (houseAge.max() - houseAge.min()) #Normalizes all the data to be between 0 and 1
```

```python
minmax.min() #Confirms that the lowest value is 0
```

```python
minmax.max() #Confirms that the highest value is 1
```

```python
minmax
```

```python
data['houseAgeStandardized'] = minmax #Appends the normalized data to the dataset.
```
